```python
import pandas as pd
```

```python
df_participate = pd.read_csv('participate_log.csv')
```

```python
df_participate
```

```python
import datetime
```

```python
# '날짜' 열을 datetime 형식으로 변환
df_participate['일자'] = pd.to_datetime(df_participate['일자'])

# 월별 열 생성
df_participate['월별'] = df_participate['일자'].dt.to_period('M')


df_participate.head()
```

```python
df_participate.groupby(by = '월별', as_index = False)['참가자 수'].mean()
```
