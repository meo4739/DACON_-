## 5. 데이콘 진행 공개 AI 경진대회 제출수 정보(submission_log.csv) 시각화
### a. 시간에 따른 총 제출 수의 변화:
- 첫 번째 그래프는 전체 기간 동안 날짜별로 총 제출 수의 변화를 보여줍니다.
- 이를 통해 특정 시기에 대회 참여와 제출 활동이 어떻게 변화하는지 확인할 수 있습니다.

### b. 특정 대회의 제출 수 추이 (대회 시작 후 경과일 기준):
- 두 번째 그래프는 선택된 몇 개의 대회에 대한 대회 시작 후 경과일별 제출 수의 변화를 보여줍니다.
- 각 대회의 제출 패턴을 비교하면서 대회의 인기와 참가자들의 활동 패턴을 이해할 수 있습니다.

```python
# '일자' 칼럼을 datetime 형식으로 변환합니다.
submission_log_df['일자'] = pd.to_datetime(submission_log_df['일자'])

# 각 대회의 시작일을 찾기 위해, 각 대회별로 가장 빠른 날짜를 찾습니다.
competition_start_dates = submission_log_df.groupby('ID')['일자'].min()

# 각 로그의 일자에서 해당 대회의 시작일을 빼서 경과일을 계산합니다.
# 이는 각 제출이 대회 시작 후 몇 번째 날에 이루어졌는지 나타냅니다.
submission_log_df['경과일'] = submission_log_df.apply(lambda row: (row['일자'] - competition_start_dates[row['ID']]).days, axis=1)

# 시각화를 위한 설정
plt.figure(figsize=(15, 12))

# 1. 시간에 따른 총 제출 수의 변화
# 날짜별로 데이터를 그룹화하고, 각 날짜의 제출 수를 합산합니다.
daily_submissions = submission_log_df.groupby('일자')['제출 수'].sum()
plt.subplot(2, 1, 1)  # 첫 번째 그래프 위치 지정
plt.plot(daily_submissions.index, daily_submissions.values)  # 날짜별 총 제출 수 그래프
plt.title('시간에 따른 총 제출 수 변화')  # 그래프 제목
plt.xlabel('날짜')  # x축 레이블
plt.ylabel('제출 수')  # y축 레이블
plt.xticks(rotation=45)  # x축 레이블 회전

# 2. 특정 대회의 제출 수 추이 (대회 시작 후 경과일 기준)
plt.subplot(2, 1, 2)  # 두 번째 그래프 위치 지정
sample_competitions = ['ID_126', 'ID_131', 'ID_147'] # 특정 샘플 대회 선정
for comp_id in sample_competitions:
    comp_data = submission_log_df[submission_log_df['ID'] == comp_id]
    plt.plot(comp_data['경과일'], comp_data['제출 수'], label=comp_id)  # 각 대회별 경과일에 따른 제출 수 그래프
plt.title('특정 대회의 경과일에 따른 제출 수 추이')  # 그래프 제목
plt.xlabel('대회 시작 후 경과일')  # x축 레이블
plt.ylabel('제출 수')  # y축 레이블
plt.legend()  # 범례 표시

plt.tight_layout()  # 그래프 레이아웃 조정
plt.show()  # 그래프 표시
```

## 8. 데이콘 진행 공개 AI 경진대회 코드 공유 관련 정보(codeshare_log.csv) 시각화
### a. 클러스터링 분석
- 클러스터링 분석을 진행하기 위해, codeshare_log.csv 데이터 내에서 코드 공유 활동 지표를 기반으로 클러스터를 형성할 것입니다. 
- 클러스터링은 비슷한 패턴을 가진 대회들을 그룹화하는 데 도움이 됩니다. 이를 위해 K-평균 클러스터링 알고리즘을 사용할 것입니다.
- 먼저 데이터를 준비하고, 적절한 클러스터 수를 결정하기 위해 엘보우 방법을 사용하여 클러스터링을 시도하겠습니다. 그런 다음, 형성된 클러스터를 시각화하겠습니다.

### b. 클러스터링 분석을 통해 codeshare_log.csv 데이터 내의 코드 공유 활동에 대한 다음과 같은 인사이트를 얻을 수 있습니다:
- 활동 패턴의 그룹화: 코드 공유 수, 댓글 수, 조회 수, 공유 수, 다운로드 수, 좋아요 수와 같은 여러 활동 지표들을 바탕으로 비슷한 패턴을 보이는 대회들을 그룹화합니다. 이는 대회 간 활동 패턴의 유사성을 파악하는 데 도움이 됩니다.
- 대회별 특성 이해: 클러스터링을 통해 각 그룹의 특성을 이해할 수 있습니다. 예를 들어, 어떤 그룹은 높은 조회 수와 다운로드 수를 보이는 반면, 다른 그룹은 높은 댓글 수와 좋아요 수를 보일 수 있습니다. 이러한 패턴은 대회의 성격이나 참가자의 행동을 이해하는 데 도움이 됩니다.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# 클러스터링에 사용할 피처 선택
features = ['코드 공유 수', '댓글 수', '조회 수', '공유 수', '다운로드 수', '좋아요 수']
X = codeshare_log_df[features]

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 엘보우 방법을 사용하여 최적의 클러스터 수 찾기
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# 엘보우 그래프 시각화
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('엘보우 방법을 사용한 최적의 클러스터 수 결정')
plt.xlabel('클러스터 수')
plt.ylabel('Inertia')
plt.show()
```

- 엘보우 방법을 사용하여 클러스터 수를 결정하는 그래프를 시각화했습니다. 이 그래프는 클러스터 수에 따른 이너셔(inertia)를 보여주며, 이너셔가 급격히 감소하는 지점을 찾아 최적의 클러스터 수를 결정하는 데 도움이 됩니다. 그래프에서 '엘보우(꿇음)' 지점이 최적의 클러스터 수를 나타냅니다. 베이스라인에서는 임의로 K=5를 설정하여, 클러스터링 분석을 진행하겠습니다.

- 이를 위해 우선 클러스터링을 진행하고, 이런 클러스터링이 전체 데이터를 잘 군집화했는지 차원축소 기법과 3D Scatterplot를 통해 확인해보겠습니다.

```python
features = ['코드 공유 수', '댓글 수', '공유 수', '다운로드 수', '좋아요 수', '조회 수']
X = codeshare_log_df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-평균 클러스터링 수행
kmeans = KMeans(n_clusters=5, random_state=0)
clusters = kmeans.fit_predict(X_scaled)

# PCA를 통한 차원축소
from sklearn.decomposition import PCA
pca = PCA(n_components=3) # 전체 feature는 6차원이여서 시각화가 힘들기 때문에 시각화를 위해 3차원으로 차원축소
pca_transformed = pca.fit_transform(X) # PCA를 통해 전체 feature들의 특징을 반영한 3개의 feature 생성

X['pca_x'] = pca_transformed[:,0] # PCA로 생성된 첫 번째 feature를 임의로 x축으로 설정
X['pca_y'] = pca_transformed[:,1] # PCA로 생성된 첫 번째 feature를 임의로 y축으로 설정
X['pca_z'] = pca_transformed[:,2] # PCA로 생성된 첫 번째 feature를 임의로 z축으로 설정

X['Cluster'] = clusters # 군집화를 통해 구분된 군집을 추가

from mpl_toolkits.mplot3d import Axes3D # 3차원 그래프 생성을 위한 패키지를 불러옵니다
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d') # 불러온 Axes3D를 fig에 추가해 3차원 시각화 공간을 형성합니다

x = X['pca_x']
y = X['pca_y']
z = X['pca_z']

ax.set_xlabel("PCA X")
ax.set_ylabel("PCA Y")
ax.set_zlabel("PCA Z")

ax.scatter(x, y, z, c= X["Cluster"]) # 3차원 Scatterplot을 그리며, 이를 X의 Cluster 값들로 구분해줍니다 

plt.show()
```

- 그 결과 각 클러스터가 전체 데이터를 잘 군집화하고 있는 것을 확인할 수 있습니다. 그럼 이제 이 클러스터별로 어떤 특징이 있는지 막대 그래프를 통해 시각화하겠습니다 

```python
# 클러스터 결과를 원본 데이터에 추가
codeshare_log_df['Cluster'] = clusters

# 클러스터별 평균값 계산
cluster_means = codeshare_log_df.groupby('Cluster').mean()

# 조회 수의 스케일을 조정합니다 (예: 천 단위로 변환)
cluster_means['조회 수 (백 단위)'] = cluster_means['조회 수'] / 100

# 수정된 조회 수와 나머지 지표들을 모두 막대 그래프로 표현합니다.
modified_features = ['코드 공유 수', '댓글 수', '공유 수', '다운로드 수', '좋아요 수', '조회 수 (백 단위)']
cluster_means[modified_features].plot(kind='bar', figsize=(15, 6))
plt.title('클러스터별 평균 활동 (조회 수 백 단위 조정)')
plt.ylabel('평균 값')
plt.xlabel('클러스터')
plt.xticks(rotation=0)
plt.show()
```
